```python
from bokeh.plotting import ColumnDataSource, figure, output_file
from bokeh.models import Slider, Select
from bokeh.io import curdoc, show 
from bokeh.layouts import widgetbox, column
from bokeh.models.tools import HoverTool

import pandas as pd
import numpy as np

from sklearn.linear_model import LinearRegression
from sklearn.datasets import load_iris
```

```python
df = pd.DataFrame(load_iris().data, columns=[load_iris().feature_names])
df['species'] = load_iris().target
```

```python
df.replace([0, 1, 2], load_iris().target_names, inplace=True)
```

```python
df.head(5)
```

```python
# create dropdowns

# Create two dropdown Select widgets: select1, select2
select_plot = Select(title='Plot', options=['Bar Plot', 'Histogram', 'Scatter Plot', 'Box Plot'],
                    value='Scatter Plot')

select_x = Select(title='X-axis', options=['sepal length (cm)', 'sepal width (cm)',
                                          'petal length (cm)', 'petal width (cm)'],
                 value='sepal length (cm)')
select_y = Select(title='Y-axis', options=['sepal length (cm)', 'sepal width (cm)',
                                          'petal length (cm)', 'petal width (cm)'],
                 value='sepal width (cm)')

# Define a callback function: callback
def update_axis(attr, old, new):
    x = df[select_x.value]
    y = df[select_y.value]
    update_plot()

def update_plot(attr, old, new):
    plot_scatter()


# Attach the callback to the 'value' property of select1
select_x.on_change('value', update_axis)
select_y.on_change('value', update_axis)
select_plot.on_change('value', update_plot)
# Create layout and add to current document
layout = widgetbox(select_plot, select_x, select_y)
curdoc().add_root(layout)

plot_scatter()
```

```python
# define plotting functions

def plot_scatter():
    x = select_x.value
    y = select_y.value
    p = figure(plot_width=500, plot_height=400)
    p.circle(df[x], df[y], size=15, fill_color="blue", fill_alpha=0.66)
    p.xaxis.axis_label = x
    p.axis.major_label_text_color = "red"
    show(p)

def plot_hist():
    x = select_x.value
    y = select_y.value
    

def plot_bar():
    x = select_x.value
    y = select_y.value

def box_plot():
    x = select_x.value
    
    y = select_y.value
```
